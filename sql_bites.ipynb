## Using COUNT, AVG, etc. with GROUP BY ##

Find the average ranking of songs per artist.

```python
%%sql

SELECT artist, 
ROUND(AVG(year_rank), 1) AS avg_rank 
FROM billboard_top_100_year_end 
GROUP BY artist 
ORDER BY avg_rank ASC
LIMIT 10
```

Find the total number of unique artists in the dataset.

```python
%%sql

SELECT COUNT(DISTINCT artist) FROM billboard_top_100_year_end
```

Find the highest and lowest ranking songs per year.

```python
%%sql

SELECT song_name, year, year_rank FROM billboard_top_100_year_end 
WHERE year_rank = 1 
OR year_rank = 100
LIMIT 10
```

## Subqueries ##

Find all songs by the artist with the most songs in the dataset.

```python
%%sql

SELECT song_name FROM billboard_top_100_year_end WHERE
artist = (
    SELECT artist
    FROM (
        SELECT artist, 
        COUNT(song_name) 
        AS song_count
        FROM(
            SELECT DISTINCT ON (song_name) artist, 
            song_name
            FROM billboard_top_100_year_end 
            ) subquery
        GROUP BY artist
        ORDER BY song_count DESC
        LIMIT 1
    ) top_artist
) ORDER BY song_name
LIMIT 10
```

Find the best-ranked song per artist

```python
%%sql

SELECT artist, song_name, year_rank
FROM billboard_top_100_year_end
WHERE (artist, year_rank) IN (
SELECT artist, MIN(year_rank) 
FROM billboard_top_100_year_end
GROUP BY artist
)
LIMIT 10
```

Find songs ranked higher than the artistâ€™s average rank.

```python
%%sql

SELECT song_name, artist, year, year_rank
FROM billboard_top_100_year_end b
WHERE year_rank < (
    SELECT AVG(year_rank)
    FROM billboard_top_100_year_end 
    WHERE artist = b.artist
);
```

## Using HAVING ##

Find artists that had more than 3 songs on the Billboard Top 100.

```python
%%sql

SELECT artist FROM billboard_top_100_year_end
GROUP BY artist
HAVING COUNT(*) > 3
LIMIT 10
```

## Window Functions ##

Use a window function to find the rolling average of the artist with the most appearences in the dataset, and describe how their average rank trends over time.

```python
%%sql

-- find the artist with the most appearances
-- display their song, year year rank
-- calculate the average year rank across all rows
-- convert this to rolling average

SELECT 
song_name,
year, 
year_rank, 
ROUND(AVG(year_rank) 
OVER (ORDER BY year ROWS BETWEEN 5 PRECEDING AND CURRENT ROW), 1) 
AS rolling_avg_rank_over_5_years
FROM billboard_top_100_year_end
WHERE artist = (
    SELECT artist
    FROM billboard_top_100_year_end
    GROUP BY artist
    ORDER BY COUNT(*) DESC
    LIMIT 1
)
ORDER BY year ASC
LIMIT 10

```
