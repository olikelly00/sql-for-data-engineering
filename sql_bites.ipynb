```python
%%sql

SELECT *, ("High Price" - "Low Price") / 2 + "Low Price" AS "Midpoint Price" FROM us_pumpkin_prices LIMIT 3;
```

```python
%%sql 

SELECT *, ("High Price" - "Low Price") AS "Price Range" FROM us_pumpkin_prices LIMIT 3;
```

```python
%%sql

-- Show three columns: city name, variety
-- and an agregated column counting the 
-- total number of rows per combination of city name and variety

SELECT "City Name", "Variety", COUNT(*)
FROM us_pumpkin_prices

-- Group this data by City Name first, then by Variety within each City Name
-- ie. we’ll get separate counts for each pumpkin variety in each city.

GROUP BY "City Name", "Variety"

-- Filter the data so that only the groups where the maximum "High Price" 
-- exceeds 300 are shown

HAVING MAX("High Price") > 300

-- Order results alphabetically by City Name A-Z

ORDER BY "City Name" ASC
```

```python
%%sql

SELECT "City Name", "Variety" FROM us_pumpkin_prices 
WHERE "Low Price" >= 5
GROUP BY "City Name", "Variety"
ORDER BY "Variety" 
LIMIT 20
```

```python
%%sql

SELECT artist AS "Artists", COUNT(*) AS "Hits" 
FROM billboard_top_100_year_end
GROUP BY artist
ORDER BY "Hits" DESC
LIMIT 10
```

```python
%%sql

SELECT "City Name", AVG("High Price") - AVG("Low Price") AS "Difference"
FROM us_pumpkin_prices
GROUP BY "City Name"
ORDER BY "Difference" DESC
```
